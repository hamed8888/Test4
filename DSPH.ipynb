```python
#imports
import pandas as pd
import json
file = 'sleep-2022-11-22.json'
with open(file) as sleepData:
    jsonVal = json.load(sleepData)
# read nested json
sleep = pd.json_normalize(jsonVal) 
data = pd.json_normalize(jsonVal,record_path =['levels','data'], meta=['logId','dateOfSleep'])
shortData = pd.json_normalize(jsonVal,record_path =['levels','shortData'], meta=['logId','dateOfSleep'])
#extract main data and details data
sleep['dateOfSleep']= pd.to_datetime(sleep['dateOfSleep'])
data['dateOfSleep']= pd.to_datetime(data['dateOfSleep'])
shortData['dateOfSleep']= pd.to_datetime(shortData['dateOfSleep'])
#extract main data columns
sleepColumns = []
for col in sleep.columns:
    sleepColumns.append(col)
print(sleepColumns)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import Range1d
import panel as pn
pn.extension()
# output to static HTML file
#output_file("line.html")
def drawPlot(column):
    p = figure(sizing_mode="stretch_width",toolbar_location="below" ,x_axis_type="datetime",width = 1000)
    p.y_range = Range1d(0, 600)
    # add a circle renderer with a size, color, and alpha
    p.circle(sleep["dateOfSleep"], sleep[column],legend_label=column, size=8, color="green", alpha=0.5)
    p.line(sleep["dateOfSleep"], sleep[column],legend_label=column, line_width=2, color="green", alpha=0.5)
    p.line(sleep["dateOfSleep"], sleep["timeInBed"],legend_label="timeInBed", line_width=2, color="blue", alpha=0.5)
    p.circle(sleep["dateOfSleep"], sleep["timeInBed"],legend_label="timeInBed", size=8, color="blue", alpha=0.5)

    p.legend.location = "top_right"
    # show the results
    return p
```

```python
select = pn.widgets.Select(name='Select', options=sleepColumns)
```

```python
pn.extension()
pn.interact(drawPlot, column= select)
```
